```python
import os
import pandas as pd
import numpy as np
import csv

path = 'files_hw/'
# Чтение файлов и папок
for file in os.listdir('files_hw/'):
    print(file)
```

```python
city_weather = path + 'city_weather.csv'
city_weather_mapper = path + 'city_weather_mapper.csv'
district_weather_mapper = path + 'district_weather_mapper.csv'
district_to_city = path + 'district_to_city.csv'
city_weather_sorted = path + 'city_weather_sorted.csv'
district_weather_sorted = path + 'district_weather_sorted.csv'
city_weather_reducer = path + 'city_weather_reducer.csv'
district_weather_reducer = path + 'district_weather_reducer.csv'
```

```python
df = pd.read_csv(district_to_city)
df.head()
```

```python
def mapper_cities(row:list) -> list:
    avg_temp = (float(row[2]) + float(row[3]))/2
    district = df[df['city'] == row[0]]['federal_district'].values[0]
    return [row[0], district, row[1], avg_temp]

def mapper_district(row:list) -> list:
    avg_temp = (float(row[2]) + float(row[3]))/2
    district = df[df['city'] == row[0]]['federal_district'].values[0]
    return [district, row[1], avg_temp]
```

```python
with open(city_weather_mapper, 'w', newline='') as w:
    writer = csv.writer(w, delimiter=',')
    with open(city_weather, 'r') as f:
        for i, line in enumerate(f):
            if i == 0:
                continue
            # print(line.strip().split(","))
            line = line.strip().split(",")
            writer.writerow(mapper_cities(line))

with open(district_weather_mapper, 'w', newline='') as w:
    writer = csv.writer(w, delimiter=',')
    with open(city_weather, 'r') as f:
        for i, line in enumerate(f):
            if i == 0:
                continue
            # print(line.strip().split(","))
            line = line.strip().split(",")
            writer.writerow(mapper_district(line))            
```

```python
print("Расчёт средней температуры в городе")

with open(city_weather_reducer, 'w', newline='') as w:
    writer = csv.writer(w, delimiter=',')
    with open(city_weather_sorted, 'r') as f:
        prev_city = None
        prev_district = None
        temp_sum = 0
        cnt = 0
        avg_temp = None
        list_for = list(enumerate(f))
        length = len(list_for)
        for i, line in list_for:
            line = line.strip().split(",")
            if prev_city == None:
                prev_city = line[0]
                prev_district = line[1]
                cnt = 1
                temp_sum = float(line[3])
            elif prev_city == line[0] and i < length - 1:
                cnt += 1
                temp_sum += float(line[3])
            elif prev_city != line[0] or i == length - 1:
                avg_temp = temp_sum/cnt
                writer.writerow([prev_city, prev_district, avg_temp])
                print([prev_city, prev_district, avg_temp])
                prev_city = line[0]
                prev_district = line[1]
                temp_sum = 0
                cnt = 0

print()
print("Расчёт средней температуры в федеральном округе")

with open(district_weather_reducer, 'w', newline='') as w:
    writer = csv.writer(w, delimiter=',')
    with open(district_weather_sorted, 'r') as f:
        prev_district = None
        temp_sum = 0
        cnt = 0
        avg_temp = None
        list_for = list(enumerate(f))
        length = len(list_for)
        for i, line in list_for:
            line = line.strip().split(",")
            if prev_district == None:
                prev_district = line[0]
                cnt = 1
                temp_sum = float(line[2])
            elif prev_district == line[0] and i < length - 1:
                cnt += 1
                temp_sum += float(line[2])
            elif prev_district != line[0] or i == length - 1:
                avg_temp = temp_sum/cnt
                writer.writerow([prev_district, avg_temp])
                print([prev_district, avg_temp])
                prev_district = line[0]
                temp_sum = 0
                cnt = 0
```

```python
print("Расчёт разницы между средней температурой города и федерального округа, к которому относится этот город")

df = pd.read_csv(district_weather_reducer, header=None)

with open(path + "city_diff_weather_reducer.csv", 'w', newline='') as w:
    writer = csv.writer(w, delimiter=',')
    with open(city_weather_reducer, 'r') as f:
        for i, line in enumerate(f):
            line = line.strip().split(",")
            district_temp = df[df[0] == line[1]][1].values[0]            
            writer.writerow([line[0], line[1], float(line[2]) - float(district_temp)])
            print([line[0], line[1], float(line[2]) - float(district_temp)])
```
